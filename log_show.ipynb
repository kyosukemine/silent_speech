```python
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib widget

val_loss = []
train_loss = []
path = "models/transduction_model_kyosuke_2400/log.txt"
path = "models/transduction_model_finetuned/log.txt"
with open(path, "r") as f:
    log = f.readlines()
    # print(log[5:2405])
    for line in log[5:205]:
        _log = line.split(" ")
        val_loss.append(float(_log[6]))
        train_loss.append(float(_log[9]))
        # print(line.split(" "))
# plt.legend()
plt.plot(train_loss)
plt.ylabel("loss")
plt.xlabel("epoch")
plt.plot(val_loss)
plt.legend(["train", "val"])
# print(path.split()[0])
plt.savefig(path.split(".")[0] + ".png")
# plt.show()





```
